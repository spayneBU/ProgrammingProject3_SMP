```python
import numpy as np
import csv
```

#### Problem 1

```python
def AfricaFile():
    '''AfricaFile - Accepts the filename of data about African country populations for 5 years and returns the dataset as a list'''
    with open('africa_pop_data.csv',newline='') as af:
        reader = csv.reader(af)
        africa_list = list(reader)
    return africa_list
```

#### Problem 2

```python
def NumData(file=AfricaFile()):
    '''NumData - Accepts a dataset (list) that contains the African country population and returns a NumPy array with only the numerical data'''
    af = np.empty((len(file),len(file[0])-2),int)
    for i in range(len(file)):
        for j in range(len(file[0])):
            try:
                if type(int(file[i][j]))==int:
                    af[i][j-2] = int(file[i][j])
            except:
                None
    return af
```

#### Problem 3

```python
from prettytable import PrettyTable
p = PrettyTable()   
```

```python
def AfricaTable(file=AfricaFile()):
    '''AfricaTable - Accepts a dataset (list) that contains the African country populations and returns a PrettyTable table of the data'''
    p.field_names = file[0]
    for i in range(1,len(file)):
        p.add_row(file[i])
    for j in range(len(file[0])-1):
        if j!=1:
            p.align[file[0][j]] = 'r'
        else:
            p.align[file[0][j]] = 'c'
    print(p)
```

#### Problem 4

```python
def SummaryTable(file=NumData()):
    '''SummaryTable - Accepts the NumPy array that contains just the population data for the African countries and that prints a summary table of statistics for each year for all African countries'''
    q = PrettyTable() 
    q.field_names = ['Summary Measure'] + [str(i) for i in list(file[0])]
    q.add_row(['MEAN',
               f'{np.mean(file[1:,0]):,.2f}',
               f'{np.mean(file[1:,1]):,.2f}',
               f'{np.mean(file[1:,2]):,.2f}',
               f'{np.mean(file[1:,3]):,.2f}',
               f'{np.mean(file[1:,4]):,.2f}'
              ])
    q.add_row(['MAX',
               f'{np.max(file[1:,0]):,.2f}',
               f'{np.max(file[1:,1]):,.2f}',
               f'{np.max(file[1:,2]):,.2f}',
               f'{np.max(file[1:,3]):,.2f}',
               f'{np.max(file[1:,4]):,.2f}'
              ])
    q.add_row(['MIN',
               f'{np.min(file[1:,0]):,.2f}',
               f'{np.min(file[1:,1]):,.2f}',
               f'{np.min(file[1:,2]):,.2f}',
               f'{np.min(file[1:,3]):,.2f}',
               f'{np.min(file[1:,4]):,.2f}',
              ])
    q.align = 'r'
    print('AFRICAN POPULATION STATISTICS'); print(q)
```

#### Problem 5
